```python
import json
import folium
import warnings
import numpy as np
import pandas as pd

warnings.filterwarnings("ignore")
```

Zillow Home Value Index (ZHVI): A smoothed, seasonally adjusted measure of the median estimated home value across a given region and housing type.

https://www.zillow.com/research/data/

```python
with open("data/dc_zip_geo.json") as jsonFile:
    geojson = json.load(jsonFile)
```

```python
geojson
```

```python
dc_zips = []
for feat in geojson['features']:
    dc_zips.append(feat["properties"]["ZCTA5CE10"])
```

```python
zip_df = pd.read_csv("data/Zip_Zhvi_AllHomes.csv", encoding='latin')#.dropna()
```

```python
dc_data = zip_df[zip_df.RegionName.isin(dc_zips)]
dc_data.rename(columns={"RegionName": "ZCTA5CE10"}, inplace=True)
dc_data.iloc[:, 1] = dc_data['ZCTA5CE10'].astype(str)
```

```python
m = folium.Map(location=[ 38.9072, -77.0369], zoom_start=11.5, tiles="Stamen Toner")

dc_json = "https://raw.githubusercontent.com/OpenDataDE/State-zip-code-GeoJSON/master/dc_district_of_columbia_zip_codes_geo.min.json"

folium.Choropleth(
    geo_data=dc_json,
    name="DC",
    data=dc_data,
    columns=["ZCTA5CE10", "2019-01"],
    key_on="feature.properties.ZCTA5CE10",
    fill_color="YlGn",
    nan_fill_color="black",
    fill_opacity=0.5,
    line_opacity=0.4,
    legend_name='Average Home Price'
).add_to(m)

folium.LayerControl().add_to(m)

m
```
